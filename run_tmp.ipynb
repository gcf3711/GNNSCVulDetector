````python
## 生成 reentrancy_contract_name.txt

import os

def export_sol_file_names(folder_path, output_file):
    num = 0
    with open(output_file, 'w') as f:
        # 遍历文件夹中的所有文件
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # 检查文件是否是 .sol 文件
                if file.endswith('.sol'):
                    # 将文件名写入到输出文件中
                    f.write(file + '\n')
                    num = num + 1
    print(num)

# 指定要遍历的文件夹路径
folder_path = './mainnet_sample_remove_comment/'

# 指定输出文件路径
output_file = './tools/reentrancy/mainnet_reentrancy_contract_name.txt'

# 调用函数将 .sol 文件名导出到输出文件中
export_sol_file_names(folder_path, output_file)

````
